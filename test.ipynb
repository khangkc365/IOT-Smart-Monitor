```python
import cv2
import matplotlib.pyplot as plt

# Đường dẫn đến ảnh trong máy
image_path = r"C:\Users\ADMIN\Downloads\landscape-responsive-card-main - home\landscape-responsive-card-main\assets\img\phongkhach.jpg"

# Đọc ảnh
image = cv2.imread(image_path)

# Kiểm tra ảnh có tồn tại không
if image is None:
    print("Không thể đọc ảnh. Kiểm tra lại đường dẫn.")
    exit()

# Làm mờ ảnh bằng Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (51, 51), 0)

# Chuyển ảnh từ BGR sang RGB để hiển thị đúng màu với matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
blurred_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)

# Hiển thị ảnh gốc và ảnh làm mờ trong cùng một cửa sổ matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Ảnh gốc')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blurred_rgb)
plt.title('Ảnh làm mờ')
plt.axis('off')

plt.tight_layout()
plt.show()
```
