```python
import cv2
import matplotlib.pyplot as plt

# ng dn n nh trong my
image_path = r"C:\Users\ADMIN\Downloads\landscape-responsive-card-main - home\landscape-responsive-card-main\assets\img\phongkhach.jpg"

# c nh
image = cv2.imread(image_path)

# Kim tra nh c tn ti khng
if image is None:
    print("Khng th c nh. Kim tra li ng dn.")
    exit()

# Lm m nh bng Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (51, 51), 0)

# Chuyn nh t BGR sang RGB  hin th ng mu vi matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
blurred_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)

# Hin th nh gc v nh lm m trong cng mt ca s matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('nh gc')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blurred_rgb)
plt.title('nh lm m')
plt.axis('off')

plt.tight_layout()
plt.show()
```
